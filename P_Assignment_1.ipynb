```python
import numpy as np
import pandas as pd
import cv2
from scipy import stats
from sklearn.model_selection import train_test_split

# 1. Basic Python Operations
# i. Assign any value to a variable and print the value and its type.
x = 10
print("Value:", x)
print("Type:", type(x))

# ii. Assign values to two variables and perform the following operations
a = 20
b = 5
print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)

# iii. Assign any string to a variable and print the length of the string.
s = "Hello, World!"
print("String:", s)
print("Length:", len(s))

# iv. Assign a value as string to a variable and then change the type of that variable to integer.
str_value = "123"
int_value = int(str_value)
print("String value:", str_value)
print("Integer value:", int_value)
print("Type of integer value:", type(int_value))

# 2. Swapping Variables
# With using a temporary variable
x = 5
y = 10
temp = x
x = y
y = temp
print("After swapping with temp variable: x =", x, "y =", y)

# Without using a temporary variable
x = 5
y = 10
x, y = y, x
print("After swapping without temp variable: x =", x, "y =", y)

# 3. String Operations
str = "I am a computer science student"

# i. Prints the first character of the string.
print("First character:", str[0])

# ii. Prints the character starting from 3rd to 5th.
print("Characters from 3rd to 5th:", str[2:5])

# iii. Prints the character starting from the 3rd character.
print("Characters from 3rd onward:", str[2:])

# 4. List Operations
# i. Create a list of 6 items and display it.
my_list = [1, 2, 3, 4, 5, 6]
print("List:", my_list)

# ii. Print all elements of the list from 3rd to the end of the list
print("Elements from 3rd to end:", my_list[2:])

# iii. Insert an element 100 to the middle of the list and also append an element 8 to the end of the list.
my_list.insert(len(my_list)//2, 100)
my_list.append(8)
print("Modified list:", my_list)

# iv. Print the 3rd, 4th and 5th element of the list.
print("3rd, 4th, 5th elements:", my_list[2:5])

# 5. Merging Lists and Printing Even Numbers
list1 = [1, 2, 3, 4]
list2 = [5, 6, 7, 8]
merged_list = list1 + list2
print("Merged list:", merged_list)
even_numbers = [x for x in merged_list if x % 2 == 0]
print("Even numbers:", even_numbers)

# 6. Tuple Operations
# i. Create a tuple of mixed data type and print it.
mixed_tuple = (1, "hello", 3.14, True)
print("Mixed tuple:", mixed_tuple)

# ii. Create a tuple with the repetition of the word “programming” 3 times.
repeated_tuple = ("programming",) * 3
print("Repeated tuple:", repeated_tuple)

# iii. Create two tuples and concatenate both the tuples.
tuple1 = (1, 2, 3)
tuple2 = (4, 5, 6)
concatenated_tuple = tuple1 + tuple2
print("Concatenated tuple:", concatenated_tuple)

# 7. Dictionary Operations
person = {'first_name': 'John', 'last_name': 'Doe', 'age': 25, 'favorite_colors': ['blue', 'green'], 'active': True}

# i. Print the dictionary.
print("Person dictionary:", person)

# ii. Print the keys and values of the dictionary separately.
print("Keys:", person.keys())
print("Values:", person.values())

# iii. Print the value for the 2nd key.
print("Value for the 2nd key:", list(person.values())[1])

# 8. Check Even or Odd
number = int(input("Enter a number: "))
if number % 2 == 0:
    print("The number is even.")
else:
    print("The number is odd.")

# 9. Display Numbers from a List Using For Loops
numbers = [1, 2, 3, 4, 5]
for number in numbers:
    print(number)

# 10. Function to Return Unique Elements
def unique_elements(lst):
    return list(set(lst))

print(unique_elements([1, 2, 2, 3, 4, 4, 5]))

# 11. Array Operations
array = np.array([[1, 6, 7, 9], [7, 9, 3, 5]])

# i. Dimension of the array
print("Dimension of the array:", array.ndim)

# ii. Shape of the array
print("Shape of the array:", array.shape)

# iii. Size of the array
print("Size of the array:", array.size)

# 12. Random Array Operations
array1 = np.random.rand(3, 3)
array2 = np.random.rand(3, 3)

# i. Concatenate two arrays.
concatenated = np.concatenate((array1, array2))
print("Concatenated array:\n", concatenated)

# ii. Sort both the arrays.
print("Sorted array1:\n", np.sort(array1, axis=None))
print("Sorted array2:\n", np.sort(array2, axis=None))

# iii. Add the two arrays
print("Addition of arrays:\n", array1 + array2)

# iv. Subtract the two arrays
print("Subtraction of arrays:\n", array1 - array2)

# v. Multiply two arrays
print("Multiplication of arrays:\n", array1 * array2)

# vi. Divide the two arrays.
print("Division of arrays:\n", array1 / array2)

# 13. NumPy Operations
# i. Create a 4 x 5 matrix with values ranging from 1 to 10, also find the transpose of the matrix.
matrix = np.random.randint(1, 11, size=(4, 5))
print("Matrix:\n", matrix)
print("Transpose of matrix:\n", matrix.T)

# ii. Create an array of 10 zeros, 10 ones, 10 fives.
zeros = np.zeros(10)
ones = np.ones(10)
fives = np.full(10, 5)
print("Zeros:", zeros)
print("Ones:", ones)
print("Fives:", fives)

# iii. Create an array of all the even integers from 10 to 50.
evens = np.arange(10, 51, 2)
print("Even integers from 10 to 50:", evens)

# iv. Generate a random number between 0 and 1.
random_number = np.random.rand()
print("Random number between 0 and 1:", random_number)

# v. Save the matrix (generated in question iv) to a text file and load it.
np.savetxt('matrix.txt', matrix)
loaded_matrix = np.loadtxt('matrix.txt')
print("Loaded matrix:\n", loaded_matrix)

# 14. Max and Min Values from Each Feature
random_matrix = np.random.rand(10, 5)
print("Random matrix:\n", random_matrix)

# Find the maximum and minimum values from each feature (column)
max_values = random_matrix.max(axis=0)
min_values = random_matrix.min(axis=0)
print("Max values from each feature:", max_values)
print("Min values from each feature:", min_values)

# 15. Count of Patterns with Same ith Feature
random_matrix = np.random.rand(10, 5)

# Count the number of patterns having the same ith feature
for i in range(random_matrix.shape[1]):
    unique, counts = np.unique(random_matrix[:, i], return_counts=True)
    print(f"Feature {i+1} has {counts.max()} patterns with the same value.")

# 16. Read and Save Image as Matrix
# Read image
image = cv2.imread('/content/Screenshot from 2024-07-13 18-12-54.png', 0)  # Change 'path_to_image.jpg' to your image path

# Convert image to DataFrame
image_df = pd.DataFrame(image)

# Save DataFrame to .csv file
image_df.to_csv('image_matrix.csv', index=False)

# 17. Import Excel Data from CSV File
# Read the CSV file, excluding the last row and last column
image_matrix = pd.read_csv('image_matrix.csv')
image_matrix = image_matrix.iloc[:-1, :-1]
print(image_matrix)

# 18. Create Today's Date
today = pd.Timestamp.today()
print("Today's date:", today)

# 19. Dataset Operations
df = pd.read_csv('/content/salary_data.csv')

# Print the number of features and patterns
num_features = df.shape[1] - 1  # Assuming the last column is the output
num_patterns = df.shape[0]
output_range = (df.iloc[:, -1].min(), df.iloc[:, -1].max())

print("Number of features:", num_features)
print("Number of patterns:", num_patterns)
print("Range of output:", output_range)

# Split the dataset
for X in range(10, 100, 10):
    Y = 100 - X
    X_train, X_test = train_test_split(df, test_size=Y/100.0, random_state=42)
    print(f"Train/Test split: {X}%/{Y}%")

# 20. Statistical Analysis
# Assuming the dataset is loaded into a DataFrame `df`
# df = pd.read_csv('path_to_dataset.csv')  # Uncomment and change the path

# Row-wise statistics
row_mean = df.mean(axis=1)
row_median = df.median(axis=1)
row_mode = df.mode(axis=1)[0]
row_std = df.std(axis=1)

print("Row-wise mean:", row_mean)
print("Row-wise median:", row_median)
print("Row-wise mode:", row_mode)
print("Row-wise standard deviation:", row_std)

# Column-wise statistics
col_mean = df.mean(axis=0)
col_median = df.median(axis=0)
col_mode = df.mode(axis=0).iloc[0]
col_std = df.std(axis=0)

print("Column-wise mean:", col_mean)
print("Column-wise median:", col_median)
print("Column-wise mode:", col_mode)
print("Column-wise standard deviation:", col_std)

# Overall statistics
overall_mean = df.values.mean()
overall_median = np.median(df.values)
overall_mode = stats.mode(df.values, axis=None)
overall_std = df.values.std()

print("Overall mean:", overall_mean)
print("Overall median:", overall_median)
print("Overall mode:", overall_mode)
print("Overall standard deviation:", overall_std)
```
